```python
import os
import json
from PIL import Image
import torch
from torchvision import transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import wandb # Import wandb


class PolygonColorDataset:
    def __init__(self, root_dir, color_list):
        self.root_dir = root_dir
        self.color_list = color_list
        self.image_filenames = []

        input_dir = os.path.join(root_dir, 'inputs')



        json_path = os.path.join(root_dir, 'data.json')
        with open(json_path, 'r') as f:
            self.annotation_data = json.load(f)


        annotated_image_filenames = {item.get('input_polygon') for item in self.annotation_data if item.get('input_polygon')}

        for img_file in os.listdir(input_dir):
            if img_file.endswith('.png'):

                filename = img_file

                if filename in annotated_image_filenames:
                    self.image_filenames.append(filename)
                else:
                    print(f"Warning: No annotation found for input file '{filename}', skipping.")


    def __len__(self):
        return len(self.image_filenames)

    def __getitem__(self, idx):
        img_filename = self.image_filenames[idx]

        img_path = os.path.join(self.root_dir, 'inputs', img_filename)
        image = Image.open(img_path).convert("RGB")


        annotation = None

        for item in self.annotation_data:
            if item.get('input_polygon') == img_filename:
                annotation = item
                break


        if annotation is None:
             print(f"Error: Annotation not found for input file '{img_filename}' after filtering. This should not happen.")
             raise ValueError(f"Annotation not found for input file '{img_filename}'")



        color_name = annotation.get('colour')

        color_tensor = torch.zeros(len(self.color_list))
        try:
            color_index = self.color_list.index(color_name)
            color_tensor[color_index] = 1
        except ValueError:
            print(f"Warning: Color '{color_name}' from annotation not found in color_list for '{img_filename}'. Returning zero tensor.")



        output_img_filename = annotation.get('output_image')
        if output_img_filename is None:
             print(f"Error: 'output_image' not found in annotation for input file '{img_filename}'.")
             raise ValueError(f"'output_image' not found in annotation for input file '{img_filename}'")

        output_img_path = os.path.join(self.root_dir, 'outputs', output_img_filename)
        output_image = Image.open(output_img_path).convert("RGB")

        # Convert images to tensors
        transform = transforms.ToTensor()
        input_tensor = transform(image)
        output_tensor = transform(output_image)


        return input_tensor, color_tensor, output_tensor

class UNetWithCondition(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, color_dim=6):
        super(UNetWithCondition, self).__init__()

        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, kernel_size=3, padding=1),
                nn.BatchNorm2d(out_c),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, kernel_size=3, padding=1),
                nn.BatchNorm2d(out_c),
                nn.ReLU(inplace=True)
            )

        self.color_fc = nn.Linear(color_dim, 128*128)
        self.pool1 = nn.MaxPool2d(2)
        self.enc2 = conv_block(64, 128)
        self.pool2 = nn.MaxPool2d(2)
        self.enc3 = conv_block(128, 256)
        self.pool3 = nn.MaxPool2d(2)

        self.bottleneck = conv_block(256, 512)

        self.up3 = nn.ConvTranspose2d(512, 256, 2, stride=2)
        self.dec3 = conv_block(512, 256)
        self.up2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec2 = conv_block(256, 128)
        self.up1 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec1 = conv_block(128, 64)

        self.final = nn.Conv2d(64, out_channels, kernel_size=1)

    def forward(self, x, color):

        B, _, H, W = x.shape
        color_embed = self.color_fc(color).view(B, 1, H, W)
        x = torch.cat([x, color_embed], dim=1);

        e1 = self.enc1(x)
        e2 = self.enc2(self.pool1(e1))
        e3 = self.enc3(self.pool2(e2))

        b = self.bottleneck(self.pool3(e3))

        d3 = self.dec3(torch.cat([self.up3(b), e3], dim=1))
        d2 = self.dec2(torch.cat([self.up2(d3), e2], dim=1))
        d1 = self.dec1(torch.cat([self.up1(d2), e1], dim=1))

        return self.final(d1)


wandb.init(project="ayna-unet-color")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

color_list = ['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'cyan', 'magenta']
train_dataset = PolygonColorDataset( # Instantiate train_dataset
    root_dir="/content/dataset/training",
    color_list=color_list
)

val_dataset = PolygonColorDataset( # Instantiate val_dataset
    root_dir="/content/dataset/validation",
    color_list=color_list
)

model = UNetWithCondition(color_dim=len(color_list)).to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)

EPOCHS = 40

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    for x, color, y in train_loader:

        x, color, y = x.to(device), color.to(device), y.to(device)

        optimizer.zero_grad()
        outputs = model(x, color);
        loss = criterion(outputs, y);
        loss.backward()
        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(train_loader);

    wandb.log({"epoch": epoch+1, "train_loss": train_loss})
    print(f"Epoch [{epoch+1}/{EPOCHS}], Loss: {train_loss:.4f}")
```

```python
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
import numpy as np
from torchvision.transforms import ToPILImage
import torchvision.transforms.functional as F
import matplotlib.cm as cm


val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False)


data_iter = iter(val_loader)
inputs, colors, ground_truths = next(data_iter)

inputs = inputs.to(device)
colors = colors.to(device)

with torch.no_grad():
    generated_outputs = model(inputs, colors)


inputs_cpu = inputs.cpu()
ground_truths_cpu = ground_truths.cpu()
generated_outputs_cpu = generated_outputs.cpu()

to_pil = ToPILImage()


fig, axes = plt.subplots(nrows=3, ncols=inputs.size(0), figsize=(inputs.size(0) * 3, 9))


dark_purple_cmap = cm.get_cmap('Purples_r')


for i in range(inputs.size(0)):
    # Input image
    input_img = to_pil(inputs_cpu[i])
    axes[0, i].imshow(input_img)
    axes[0, i].set_title(f"Input {i+1}")
    axes[0, i].axis('off')


    ground_truth_img = to_pil(ground_truths_cpu[i])
    axes[1, i].imshow(ground_truth_img)
    axes[1, i].set_title(f"Ground Truth {i+1}")
    axes[1, i].axis('off')


    generated_img_tensor = torch.clamp(generated_outputs_cpu[i], 0, 1)


    generated_pil = to_pil(generated_img_tensor)
    generated_gray = F.to_grayscale(generated_pil)

    axes[2, i].imshow(generated_gray, cmap=dark_purple_cmap)
    axes[2, i].set_title(f"Generated (Dark Purple Colormap) {i+1}")
    axes[2, i].axis('off')

plt.tight_layout()
plt.show()
```
